```
import numpy as np
import plotly.express as px
```

```
def logistic(n0, r, itermax):
    result = []
    n = n0
    result.append(n)
    for i in range(itermax):
        n = r * n * (1-n)
        result.append(n)
    return result
```

```
n0 = 0.5
r = 3.9
itermax = 20

data = logistic(n0, r, itermax)

fig = px.line(x=range(itermax+1), y=data)
fig.show()
```

```
n0 = 0.5
itermax = 20
ry = []
rx = np.linspace(3.56, 4, num=100_000)
for r in rx:
    ry.append(logistic(n0, r, itermax)[-1])
```

```
fig = px.scatter(x=rx, y=ry)
fig.update_layout(
    width=4000,
    height=1300,
    paper_bgcolor="LightSteelBlue",
)
fig.update_traces(mode='markers', marker_line_width=1, marker_size=1)
fig.show()
```
